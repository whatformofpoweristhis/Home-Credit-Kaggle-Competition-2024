```python
import numpy as np
def calculate_slope(x, y):
    if len(x) != len(y):
        raise ValueError("The lists x and y must have the same length.")

    n = len(x)
    sum_x = sum(x)
    sum_y = sum(y)
    sum_xy = sum(x[i] * y[i] for i in range(n))
    sum_x_squared = sum(x[i] ** 2 for i in range(n))

    numerator = n * sum_xy - sum_x * sum_y
    denominator = n * sum_x_squared - sum_x ** 2

    if denominator == 0:
        raise ValueError("Denominator is zero, the data points are likely all the same or in a perfect vertical line.")

    m = numerator / denominator
    return m

def generate_weekly_auc(a: float, b: float) -> list:
    slope = (b - a) / 100
    y_values = [a + slope * x for x in range(100)]
    return y_values


def calculate_score(arr: list) -> float:
    if len(arr) != 100:
        raise ValueError("输入数组的长度必须为100")
    mean_value = sum(arr) / len(arr)
    x_coords = list(range(100))
    adjacent_diff=calculate_slope(x_coords,arr)
    residual=[]
    for i in range(100):
        residual.append(abs(arr[i]-adjacent_diff*i+arr[0]))
    result = 2*mean_value-1 + 88 * min(0,adjacent_diff) - 0.5*np.std(residual)
#     print(f"variance penality is {0.5*(variance**0.5)}")
    return result

def test(x,y,p):
    a=generate_weekly_auc(x,y)
    print(f"模型平均auc{sum(a)/len(a)}")
    b=calculate_score(a)
    print(f"不hack取得分数{b}")
    dict={}
    def opt(a):
        o=[i / 1000 for i in range(5, 125, 1)]
        for j in o:
            c=a
            for i in range(0, p):
                c[i] = c[i] - j
            b = calculate_score(c)
            dict[j]=b
        sorted_items = sorted(dict.items(), key=lambda item: item[1], reverse=True)
        top_five = sorted_items[:1]
        for k, v in top_five:
            print(f"故意将前{p}%的weekly auc减少{k}，得到分数{v}")
    opt(a)
    print('\n')
    
def test_new(x,y,p):
    a=generate_weekly_auc(x,y)
    print(f"模型平均auc{sum(a)/len(a)}")
    b=calculate_score(a)
    print(f"不hack取得分数{b}")
    dict={}
    def opt(a):
        o=[i / 1000 for i in range(0, 125, 1)]
        for j in o:
            c=a
            for i in range(0, p):
                c[i] = c[p-1] - j
            b = calculate_score(c)
            dict[j]=b
        sorted_items = sorted(dict.items(), key=lambda item: item[1], reverse=True)
        top_five = sorted_items[:1]
        for k, v in top_five:
            print(f"故意将前{p}%的weekly auc减少{k}，得到分数{v}")
    opt(a)
    print('\n')


def getxy(average_auc,non_hack_score):
    x=average_auc+0.005
    y=average_auc-0.005
    a = generate_weekly_auc(x, y)
    while abs(calculate_score(a)-non_hack_score)>0.0003:
        if calculate_score(a)>non_hack_score:
            x=x+0.00001
            y=y-0.00001
        else:
            x=x-0.00001
            y=y+0.00001
        a=generate_weekly_auc(x,y)
    return x,y
```
